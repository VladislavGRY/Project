Часто редкие значения могут появляться по нескольким причинам:

- Ошибки ввода данных: редкие значения могут быть результатом опечаток или ошибок при вводе данных.

- Непопулярные данные: редкие значения могут возникать из-за мало известных или не популярных жанров

Выбивающиеся значения могут возникать по следующим причинам:

- Ошибки сбора данных: выбивающиеся значения могут быть результатом ошибок при сборе данных, например, ошибочных измерений или записей.
- Аномальные события: выбивающиеся значения могут отражать аномальные события или редкие случаи, которые выходят за пределы обычных ожиданий.
- Некорректные данные: выбивающиеся значения могут возникать из-за некорректных или нетипичных данных, которые не соответствуют ожидаемым значениям в столбце.

**Коэффициенты корреляции** используются для измерения прочности связи между двумя переменными.<br>
Корреляция Пирсона является наиболее часто используемой в статистике. Она измеряет **силу** и **направление** линейной зависимости между двумя переменными.<br>
Значения всегда находятся **в диапазоне от -1 (сильная отрицательная связь) до +1 (сильная положительная связь)**. Значения при нулевом значении или близкие к нулю подразумевают слабую или отсутствующую связь.<br>
**Значения коэффициентов корреляции менее +0,8 или более -0,8 не считаются значимыми**.<br><br>
**Сила связи** в общей интерпретации<br>
В общем случае значения коэффициента корреляции интерпретируют так:<br>
- Если значение положительно (находится от 0 до 1), то связь между величинами положительная.<br>
- Если значение отрицательно (находится между −1 до 0), то связь между величинами отрицательная.<br>
- Чем больше абсолютное значение коэффициента корреляции, тем сильнее связь.<br>
- Если значение коэффициента корреляции Пирсона около 0, то линейной связи между величинами нет или она очень слабая.
